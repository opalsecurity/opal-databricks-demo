```python
import requests
from datetime import date, datetime, timedelta

def connect_opal(base_url, token):
    global headers, base
    headers = {"Authorization": "Bearer " + token, "Accept": "application/json"}
    base = base_url

def invoke_opal_get_method(url):
    uri = base + url
    response = requests.get(url=uri,headers=headers, timeout=(10, 300))
    return response

def invoke_opal_post_method(url, data = None):
    uri = base + url
    response = requests.post(url=uri,headers=headers, timeout=(10, 300), data=json.dumps(data))
    return response

def get_opal_group_visibility(group_id):
    ret = invoke_opal_get_method(url="/v1/groups/" + group_id + "/visibility")
    return ret

def get_opal_users(cursor, page_size):
    if cursor != "" and page_size != "":
        ret = invoke_opal_get_method(url="/v1/users?cursor="+cursor+"&page_size="+ str(page_size))
    elif cursor != "":
        ret = invoke_opal_get_method(url="/v1/users?cursor="+cursor)
    elif page_size != "":
        ret = invoke_opal_get_method(url="/v1/users?page_size="+ str(page_size))
    else:
        ret = invoke_opal_get_method(url="/v1/users")
    return ret
    
def get_opal_groups(cursor, page_size):
    if cursor != "" and page_size != "":
        ret = invoke_opal_get_method(url="/v1/groups?cursor=" + cursor + "&page_size=" + str(page_size))
    elif cursor != "":
        ret = invoke_opal_get_method(url="/v1/groups?cursor=" + cursor)
    elif page_size != "":
        ret = invoke_opal_get_method(url="/v1/groups?page_size=" + str(page_size))
    else:
        ret = invoke_opal_get_method(url="/v1/groups")
    return ret
            
def get_opal_requests(cursor='', page_size=5000, start_date='', end_date=''):
    if start_date != "" and end_date != "" and page_size != None:
        ret = invoke_opal_get_method(url="/v1/requests?start_date=" + start_date + "&end_date=" + end_date + "&page_size=" + str(page_size))
    elif cursor != "" and page_size != "":
        ret = invoke_opal_get_method(url="/v1/requests?cursor=" + cursor + "&page_size=" + str(page_size))
    elif cursor != "":
        ret = invoke_opal_get_method(url="/v1/requests?cursor=" + cursor)
    elif page_size != "":
        ret = invoke_opal_get_method(url="/v1/requests?page_size=" + str(page_size))
    else:
        ret = invoke_opal_get_method(url="/v1/requests")
    return ret

def get_opal_owners(cursor, page_size):
    if cursor != "" and page_size != "":
        ret = invoke_opal_get_method(url="/v1/owners?cursor=" + cursor + "&page_size=" + str(page_size))
    elif cursor != "":
        ret = invoke_opal_get_method(url="/v1/owners?cursor=" + cursor)
    elif page_size != "":
        ret = invoke_opal_get_method(url="/v1/owners?page_size=" + str(page_size))
    else:
        ret = invoke_opal_get_method(url="/v1/owners")
    return ret

def get_opal_owner_users(owner_id):
    ret = invoke_opal_get_method(url="/v1/owners/" + owner_id + "/users")
    return ret

def get_opal_events(cursor, page_size):
    #base_url = "/v1/events"
    if cursor != "" and page_size != "":
        full_url = "/v1/events?page_size=" + page_size + "&cursor=" + cursor
    elif cursor != "":
        full_url = "/v1/events?cursor=" + cursor
    elif page_size != "":
        full_url = "/v1/events?page_size=" + page_size
    else:
        full_url = "/v1/events"
    ret = invoke_opal_get_method(url=full_url)
    
    return ret

def get_opal_group_reviewer_stages(group_id):
    ret = invoke_opal_get_method(url="/v1/groups/" + group_id + "/reviewer-stages")
    return ret

def get_opal_group_members(group_id):
    ret = invoke_opal_get_method(url="/v1/groups/" + group_id + "/users")
    return ret

def get_opal_group(group_id = None, group_name = None):
    if group_name != None:
        ret = invoke_opal_get_method(url="/v1/groups?group_name=" + group_name)
        return ret
    else:
        ret = invoke_opal_get_method(url="/v1/groups/" + group_id)
        return ret

def get_opal_request(request_id):
    ret = invoke_opal_get_method(url="/v1/requests/" + request_id)
    return ret

def submit_opal_request(request):
    ret = invoke_opal_post_method(url="/v1/requests", data=json.dumps(request))
    return ret

def get_opal_sync_errors():
    ret = invoke_opal_get_method(url="/v1/sync_errors?app_id=[your_okta_app_id]")
    return ret
```
